```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
import math 

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

from scipy import stats
from scipy.stats import kurtosis, skew

%matplotlib inline
```

W związku z tym z czym koreluje najsilniej zmienna price,
postanowiłyśmy usunąć niepotrzebne kolumny i zostawić tylko price, sqft_living 

```python
# Załadowanie danych
price_data = pd.read_csv("price_data.csv")
#usunięcie niepotrzebnych kolumn
price_data = price_data.drop(['grade','bathrooms','bedrooms','floors','waterfront','view','condition','sqft_lot15','sqft_living15','lat','long','zipcode','yr_renovated','yr_built','sqft_basement','sqft_above','floors','sqft_lot','date','id'], axis = 1)
# pokaż pierwsze 5 wierszy
price_data.head()
```

```python
# sprawdzenie czy nie ma brakujących wartości w komórkach
display(price_data.isna().any())

# usunięcie brakujących wartości
price_data = price_data.dropna()

# sprawdzenie czy wszystko zostało usunięte
price_data.isna().any()
```

```python
# zdefiniowanie danych x & y.
x = price_data['sqft_living']
y = price_data['price']

# stworzenie wykresu.
plt.plot(x, y, 'o', color ='blue', label = 'cena')

# upewnienie się, że jest sformatowany.
plt.title("powierzchnia mieszkalna Vs. cena nieruchomości ")
plt.xlabel("powierzchnia mieszkalna")
plt.ylabel("cena nieruchomości")
plt.legend()

plt.show()
```

```python
# podsumowanie statystyczne.
price_data.describe()
```

Teraz należy podzielić dane na zestaw szkoleniowy i zestaw testowy. Zestaw treningowy jest tym, na czym będziemy trenować model, a zestaw testowy jest tym, na czym będziemy go testować.

```python
# zdefiniowanie zmiennej wejściowej (X) i zmiennej wyjściowej.
Y = price_data.drop('price', axis = 1)
X = price_data[['price']]

# Podzielenie X i y na X_
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=1)
```

```python
#stworzenie modelu regresji liniowej.
regression_model = LinearRegression()

# przekazanie przez zestaw danych X_train & y_train.
regression_model.fit(X_train, y_train)
```

```python
# testowanie prognozy
prediction = regression_model.predict([[67.33]])
predicted_value = prediction[0][0]
print("prognozowana wartość wynosi {:.4}".format(predicted_value))
```

```python

# Testowanie kilku prognoz.
y_predict = regression_model.predict(X_test)

# Pokazanie pierwszych 5 prognoz
y_predict[:5]
```

```python
# pokazanie wyników
plt.scatter(X_test, y_test,  color='pink', label = 'Price')
plt.plot(X_test, y_predict, color='yellow', linewidth = 3, linestyle= '-',label ='Regression Line')

plt.title("powierzchnia mieszkalna Vs. cena nieruchomości")
plt.xlabel("powierzchnia mieszkalna")
plt.ylabel("cena")
plt.legend()
plt.show()
```

```python
#użycie wszystkich funkcji z danych
#podział 60/40
X = df.iloc[:, :-1].values
y = df['price'].values
X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.4, random_state=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor(n_estimators=1000,
                              criterion='mse',
                              random_state=1,
                              n_jobs=1)
forest.fit(X_train, y_train)
y_train_pred = forest.predict(X_train)
y_test_pred = forest.predict(X_test)
print('R^2 train: %.3f, test: %.3f' %
     (r2_score(y_train, y_train_pred),
     r2_score(y_test, y_test_pred)))
```

Zastosowanie współczynnika determinacji w analizach:

służy określeniu, na ile poszczególne modele statystyczne, czynniki "dobrze" wyjaśniają to co chcemy wyjaśnić, która ze zmiennych (jeżeli badamy w badaniu kilka) lepiej wyjaśnia zmienną zależną
pozwala oszacować, który z analizowanych modeli jest lepszy.
